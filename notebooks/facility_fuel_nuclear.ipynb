```python
params = {'api_key' : API_KEY}

baseurl = 'https://api.eia.gov/v2/'

endpoint = 'electricity/facility-fuel'

data_end_tag = '/data?'

facets = 'frequency=monthly&data[0]=consumption-for-eg&data[1]=consumption-for-eg-btu&data[2]=generation&data[3]=gross-generation&facets[fuel2002][]=NUC&facets[primeMover][]=ALL&start=2021-01&end=2025-07&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000'

datapoint = '&data[]=value'

url = baseurl + endpoint + data_end_tag + facets

response = requests.get(url=url, params=params)

print(response.status_code)
print(url)
```

```python
data = response.json()
response.json()
```

```python
#pd.DataFrame(data['response']['routes'])

pd.DataFrame(data['response']['data'])
```

```python
#Write data to a dataframe for analysis

facility_fuel_nuclear2 = pd.DataFrame(data['response']['data'])
```

```python
## Concatenate all data frames
facility_fuel_nuclear_all = pd.concat([facility_fuel_nuclear1,facility_fuel_nuclear2], axis=0)
```

```python
#Column DataTypes Before
facility_fuel_nuclear_all.dtypes
```

```python
#Change the data types of specific columns

facility_fuel_nuclear_all['period'] = pd.to_datetime(facility_fuel_nuclear_all['period'], format = "%Y-%m")

facility_fuel_nuclear_all['consumption-for-eg'] = facility_fuel_nuclear_all['consumption-for-eg'].astype(int)

facility_fuel_nuclear_all['consumption-for-eg-btu'] = facility_fuel_nuclear_all['consumption-for-eg-btu'].astype(int)

facility_fuel_nuclear_all['generation'] = facility_fuel_nuclear_all['generation'].astype(float) # 2 step process b/c there is a decimal mixed in

facility_fuel_nuclear_all['generation'] = facility_fuel_nuclear_all['generation'].astype(int) # 2 step process b/c there is a decimal mixed in

facility_fuel_nuclear_all['gross-generation'] = facility_fuel_nuclear_all['gross-generation'].astype(int)


#Extract year and month from period

facility_fuel_nuclear_all['year'] = facility_fuel_nuclear_all['period'].dt.year
facility_fuel_nuclear_all['month'] = facility_fuel_nuclear_all['period'].dt.month
```

```python
#Create dataframe list

dataframes = [facility_fuel_nuclear_all]

#For loop to go through each dataframe in list and format all column names to lowecase

for x in dataframes:
    x.columns= x.columns.str.lower()
```
