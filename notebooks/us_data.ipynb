```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
file_path = "../data/keys.json"

with open(file_path, 'r') as file:
    credentials = json.load(file)
    
API_KEY = credentials['api_key']
```

```python
params = {'api_key' : API_KEY}

baseurl = 'https://api.eia.gov/v2/'

endpoint = 'electricity/electric-power-operational-data'

data_end_tag = '/data?'

facets = '&facets[sectorid][]=98&frequency=monthly&start=2020-01&end=2020-01'

datapoint = '&data[]=value'

#endpoint = '/electricity/retail-sales/data?&data[]=price&facets[sectorid][]=RES&facets[stateid][]=CO&frequency=monthly&start=2000-01'

url = baseurl + endpoint + data_end_tag + facets

response = requests.get(url=url, params=params)

print(response.status_code)

##api_string = '?api_key='  
##response = requests.get(baseurl + endpoint + f'{api_string}{api_key}')
##https://api.eia.gov/v2/electricity/operating-generator-capacity?api_key=xxxxxx
## url = f'https://api.eia.gov/v2/electricity/state-electricity-profiles/capability{api_string}{api_key}'
```

```python
#Create dataframe list

dataframes = [us_data]

#For loop to go through each dataframe in list and format all column names to lowecase

for x in dataframes:
    x.columns= x.columns.str.lower()
```

```python
#Change the data types of specific columns

us_data['period'] = pd.to_datetime(us_data['period'], format = "%Y-%m")

us_data['productid'] = us_data['productid'].astype(int)

us_data['activityid'] = us_data['activityid'].astype(int)

us_data['value'] = us_data['value'].fillna(0).astype(float)

us_data['value'] = us_data['value'].astype(int)

#Extract year and month from period

us_data['year'] = us_data['period'].dt.year

us_data['month'] = us_data['period'].dt.month
```
